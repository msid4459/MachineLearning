```python
#2.1 TREAT MISSING VALUES
missing=np.nan
series_obj1=Series(['row 1','row 2',missing,'row 4','row 5',missing,'row 6'])
series_obj1
```

```python
series_obj1.isnull()
```

```python
#fill in the missing values:
np.random.seed(25)
DF_obj1=DataFrame(np.random.randn(36).reshape(6,6))
DF_obj1
```

```python
DF_obj1.iloc[3:5,0]=missing
DF_obj1.iloc[1:4,5]=missing
DF_obj1
```

```python
#fillna method (3 ways to do it: by number, by passing dictionary and by ffill)
filled_DF=DF_obj1.fillna(0)
filled_DF
```

```python
#you can use dictionary to fill specific values in column
filled_DF1=DF_obj1.fillna({0:0.1, 5:1.25})
filled_DF1
```

```python
filled_DF2=DF_obj1.fillna(method='ffill')
filled_DF2
#method=ffill , fill-forward any
#missing values with values from the last non-null element in the column series.
```

```python
#Counting Missing Values
#(reminder:) once you have found missing values you can either fill them in or drop them.
np.random.seed(25)
DF_obj3=DataFrame(np.random.randn(36).reshape(6,6))
DF_obj3.iloc[3:5,0]=missing
DF_obj3.iloc[1:4,5]=missing
DF_obj3

```

```python
DF_obj3.isnull().sum()
#this returns the count for each column
# the result below specifies: column 0 has 2 missing values and
#column 5 has 3 missing values.
```

```python
#Filtering out Missing values
#dropna method drops every single row that has NaN in it.
#that's sort of excessive because you have lost most of the data
DF_no_NaN=DF_obj3.dropna()
DF_no_NaN
```

```python
#also instead of dropping rows you can drop columns that have missing values in it.
DF_no_NaN1=DF_obj3.dropna(axis=1)
DF_no_NaN1
```

```python
#To drop rows that just have missing values (all)
DF_no_NaN2=DF_obj3.dropna(how='all')
DF_no_NaN2
```

```python
#Another way to add data to a dataset is to use the append method
#this method allows you to add rows to the bottom of a table.
#say let's  append variable_added with variable_added with itself.
#when you use the append method, you always have to pass ignore_index, whether or not you want to
#pandas to reindex dataframe for you. We passed in False
added_datatable=variable_added.append(variable_added,ignore_index=False)
added_datatable
```

```python
#now let's reindex
added_datatable=variable_added.append(variable_added,ignore_index=True)
added_datatable
```
